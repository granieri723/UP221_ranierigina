```python
la_gpd_df = la_gpd_df.rename(
    columns={
              'B01001_001E_x' : 'total_pop',
              'B02001_002E_x' : 'white_pop', 
              'B02001_003E_x' : 'black_pop',
              'B02001_004E_x' : 'american_indian_pop',
              'B02001_005E_x' : 'asian_pop',
              'B03002_012E_x' : 'hispanic',
              'B03002_002E_x' : 'non-hispanic',
              'B01001_020E_x' : '65-66-male',
              'B01001_021E_x' : '67-69-male',
              'B01001_022E_x' : '70-74-male',
              'B01001_023E_x' : '75-79-male',
              'B01001_024E_x' : '80-84-male',
              'B01001_025E_x' : '85+-male',
              'B01001_044E_x' : '65-66-female',
              'B01001_045E_x' : '67-69-female',
              'B01001_046E_x' : '70-74-female',
              'B01001_047E_x' : '75-79-female',
              'B01001_048E_x' : '80-84-female',
              'B01001_049E_x' : '85+-female',
              'B28006_001E_x' : 'total education',
              'B28006_002E_x' : 'less than high school',
              'B28006_008E_x' : 'high school',
              'B16009_001E_x' : 'total poverty level',
              'B16009_002E_x' : 'below poverty level',
              'B25044_001E_x' : 'total vehicle',
              'B25044_003E_x' : 'owner occupied: no vehicle',
              'B25044_010E_x' : 'renter occupied: no vehicle',
              'B03002_003E_x' : 'non_hispanic_white',
              'B25034_001E_x' : 'total_housing',
              'B25034_008E_x' : 'housing_1960_1969',
              'B25034_009E_x' : 'housing_1950_1959',
              'B25034_010E_x' : 'housing_1940_1949',
              'B25034_011E_x' : 'housing_1939_earlier',
              'B25004_001E_x' : 'vacant_housing_units',
              'B25004_002E_x' : 'rent_housing_units',
              'B25004_004E_x' : 'sale_housing_units',
              'B25004_008E_x' : 'other_vacant_housing units',
              'B25058_001E_x' : 'median_contract_rent',
              'B25105_001E_x' : 'median_monthly_housing_costs',
              'B25106_024E_x' : 'renter_occupied_housing_units',
              'B25106_028E_x' : 'less_than_20k',
              'B25106_032E_x' : '20k_to_34k',
              'B25106_036E_x' : '35k_to_49k',
              'B25106_040E_x' : '50k_to_74k'
    })
```

```python
## calculating relevant statistics

# calculating percent of people of color
la_gpd_df['rate_poc'] = ((la_gpd_df['hispanic'] + la_gpd_df['non-hispanic'] - la_gpd_df['non_hispanic_white']) / (la_gpd_df['hispanic'] + la_gpd_df['non-hispanic'])) 

# calculating percent of white
la_gpd_df['rate_white'] = la_gpd_df['non_hispanic_white']/ (la_gpd_df['hispanic'] + la_gpd_df['non-hispanic']) 

# calculating relative frequency of old housing units
la_gpd_df['rate_old_bldg'] = (la_gpd_df['housing_1960_1969'] + la_gpd_df['housing_1950_1959'] +  la_gpd_df['housing_1940_1949'] +  la_gpd_df['housing_1939_earlier']) / la_gpd_df['total_housing']

# calculating relative frequency of 65+ age
la_gpd_df['rate_65+'] = (la_gpd_df['65-66-male'] + la_gpd_df['67-69-male'] + la_gpd_df['70-74-male'] + la_gpd_df['75-79-male'] + la_gpd_df['80-84-male'] + la_gpd_df['80-84-male'] +  la_gpd_df['85+-male'] + la_gpd_df['65-66-female'] + la_gpd_df['67-69-female'] + la_gpd_df['70-74-female'] + la_gpd_df['75-79-female'] + la_gpd_df['80-84-female'] + la_gpd_df['80-84-female'] +  la_gpd_df['85+-female'])/ (la_gpd_df['total_pop']) 

# calculating proportion of "housing burdened"
la_gpd_df["less_than_20k_prop"] = la_gpd_df["less_than_20k"] / la_gpd_df["renter_occupied_housing_units"]

# calculating less than high school sample
la_gpd_df["less_than_hs_prop"] = la_gpd_df["less than high school"] / la_gpd_df["total education"]

# calculating below poverty level households
la_gpd_df["below_poverty_prop"] = la_gpd_df["below poverty level"] / la_gpd_df["total poverty level"]

# calculating no vehicle households
la_gpd_df["no_vehicle_prop"] = (la_gpd_df["owner occupied: no vehicle"] + la_gpd_df["renter occupied: no vehicle"]) / la_gpd_df["total vehicle"]
```

```python
la_gpd_df = la_gpd_df.dropna()
```

```python
classifier_rate_poc = mapclassify.NaturalBreaks(y=la_gpd_df['rate_poc'], k=5)
classifier_rate_old_bldg = mapclassify.NaturalBreaks(y=la_gpd_df['rate_old_bldg'], k=5)
classifier_rate_65 = mapclassify.NaturalBreaks(y=la_gpd_df['rate_65+'], k=5)
classifier_less_than_20k_prop = mapclassify.NaturalBreaks(y=la_gpd_df["less_than_20k_prop"], k=5)
classifier_less_than_hs_prop = mapclassify.NaturalBreaks(y=la_gpd_df["less_than_hs_prop"], k=5)
classifier_below_poverty_prop = mapclassify.NaturalBreaks(y=la_gpd_df["below_poverty_prop"], k=5)
classifier_no_vehicle_prop = mapclassify.NaturalBreaks(y=la_gpd_df["no_vehicle_prop"], k=5)
```

```python
la_gpd_df['rate_poc_class'] = la_gpd_df[['rate_poc']].apply(classifier_rate_poc)
la_gpd_df['rate_old_bldg_class'] = la_gpd_df[['rate_old_bldg']].apply(classifier_rate_old_bldg)
la_gpd_df['rate_65+_class'] = la_gpd_df[['rate_65+']].apply(classifier_rate_65)
la_gpd_df['less_than_20k_prop_class'] = la_gpd_df[['less_than_20k_prop']].apply(classifier_less_than_20k_prop)
la_gpd_df['less_than_hs_prop_class'] = la_gpd_df[['less_than_hs_prop']].apply(classifier_less_than_hs_prop)
la_gpd_df['below_poverty_prop_class'] = la_gpd_df[['below_poverty_prop']].apply(classifier_below_poverty_prop)
la_gpd_df['no_vehicle_prop_class'] = la_gpd_df[['no_vehicle_prop']].apply(classifier_no_vehicle_prop)
```

```python
la_gpd_df['vulnerability_index'] = (la_gpd_df['rate_poc_class'] +
                                    la_gpd_df['rate_old_bldg_class'] +
                                    la_gpd_df['rate_65+_class'] +
                                    la_gpd_df['less_than_20k_prop_class'] +
                                    la_gpd_df['less_than_hs_prop_class'] +
                                    la_gpd_df['below_poverty_prop_class'] +
                                    la_gpd_df['no_vehicle_prop_class']
                                   )
la_gpd_df.info()
```

```python
palisades_vulnerability = la_gpd_df.sjoin(palisades, how='right', predicate='intersects', on_attribute='GEOID')
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10,12))
palisades_vulnerability.plot(ax = ax[0],
    column = 'vulnerability_index',
    scheme = "EqualInterval",
    k = 5, 
    alpha = 0.5,
    legend = True,
    cmap = 'Reds',
    missing_kwds={
        "color": "lightgrey",
        "hatch": "///",
        "label": "Missing values"},
    legend_kwds={
        'loc':'upper left',
        'bbox_to_anchor':(1,.9)}
)
ax[0].set_title("Palisades Housing Vulnerbality")
```
